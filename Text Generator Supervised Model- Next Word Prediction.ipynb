```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequence = []
for sentence in data.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(tokenized_sentence)):
        n_gram = tokenized_sentence[:i+1]
        input_sequence.append(n_gram)
```

```python
max_length = max([len(x) for x in input_sequence])
```

```python
max_length
```

```python
padded_input_sequences = pad_sequences(input_sequence, maxlen=max_length, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[: , :-1]
X
```

```python
y= padded_input_sequences[: , -1]
y
```

```python
#multiclass classification problem, softmax
```

```python
X.shape
```

```python
y.shape
```

```python
y = to_categorical(y, num_classes=756)
```

```python
X.shape
```

```python
y.shape
```

```python
y[0]
```

```python
model = Sequential()
```

```python
model.add(Embedding(756,100,input_length = 128))
model.add(LSTM(160))
model.add(Dense(756, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = 'accuracy')
```

```python
model.summary()
```

```python
model.fit(X,y,epochs = 40)
```
